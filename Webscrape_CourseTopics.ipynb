## Webscraping Coursera - part 1

This notebook is about webscraping Coursera to find the commonly recommended course topics for the search term 'Artificial Intelligence'. It can be used to design a college program comprising of artificial intelligence courses.

```python
import sys
import datetime
import time
from time import sleep
from selenium import webdriver
import pandas as pd
import json
import requests
from collections import Counter
import matplotlib.pyplot as plt
import re
```

```python
driver = webdriver.Chrome(executable_path=r'chromedriver1.exe')
# login
driver.get(f'https://www.coursera.org/search?query=artificial%20intelligence')
driver.maximize_window()
time.sleep(2)


login= driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div/div/div[1]/div[2]/div[1]/div/div[2]/div/div[2]/div[4]/div/div/div/div/div[1]')
login.click() # Send mouse click`
login1= driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div/div/div[1]/div[2]/div[1]/div/div[2]/div/div[2]/div[4]/div/div/div[2]/div/div[10]/div/button[2]')
login1.click() # Send mouse click`
# login2= driver.find_elements_by_xpath('//*[@id="rendered-content"]/div/div/div[1]/div/div/div[1]/div[2]/div[1]/div/div[2]/div/div[2]/div[4]/div/div[2]/div/div[2]/div[2]/div[1]/div[1]/button/label/input')
Course_list = []
# for x in range(len(login2)):
Course_list1 = driver.find_elements_by_class_name('checkboxText')
Course_list = [x.text for x in Course_list1]
Course_count1 = driver.find_elements_by_class_name('filter-count')
Course_count1 = [x.text for x in Course_count1]
```

```python
tolist = []
for num in Course_count1:   
  abc = re.sub("[()]", '',num)
  tolist.append(int(abc))
```

```python
list1 = dict(zip(Course_list[:50],tolist[:50]))
```

```python
plt.figure(figsize = (20,10))
plt.bar(list1.keys(), list1.values())
plt.title('Most common course topics - Artificial Intelligence', size = 25)
plt.ylabel('Frequency', size=20)
plt.xticks(size = 15,rotation = 90)
plt.show()
```
